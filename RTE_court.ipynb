```python
def consommation(src):
    #src est un nom de fichier string
    dico={}
    dico_propre={}
    dico_date_jour={}
    dico_final={}
    dico_final_final={}
    liste_tuple=[]
    liste_clefs=[]

    book = xlrd.open_workbook(src)
    sh = book.sheet_by_index(0)
    ligne=np.arange(20,sh.nrows)
    colonne=np.arange(2,sh.ncols)
    
    for rx in ligne:
        for cl in colonne:
            dico[str(sh.cell_value(rowx=rx, colx=0))+' '+str(sh.cell_value(rowx=18, colx=cl))]=sh.cell_value(rowx=rx, colx=cl)
       
    for key in dico:
        liste_clefs.append(key)
        
    for key in liste_clefs:
        if dico[key]=='':
            dico.pop(key)
        
    for key in dico:
        dico_propre[key[:-6],key[-5:]] = dico[key]
    
    for key in dico_propre:
        date_object = datetime.datetime.strptime(key[0], '%d/%m/%Y')
        dico_date_jour[date_object,key[1]]=dico_propre[key]
    
    for key in dico_date_jour:
        if key[1][-5:-3]=='24':
            nouvelle_date=key[0]+datetime.timedelta(days=1)
            dico_final[nouvelle_date,'00:00'] = dico_date_jour[key]
        else:
            dico_final[key]=dico_date_jour[key]

    for key in dico_final:
        nouvelle_date_heure=key[0]+datetime.timedelta(hours=int(key[1][-5:-3]),minutes=int(key[1][-2:]))
        dico_final_final[nouvelle_date_heure] = dico_final[key]
    
    for key in dico_final_final:
        liste_tuple.append((key,dico_final_final[key]))
    
    return liste_tuple
```

```python
tuple_annee=[]
for fichier in f[2]:
    tuple_annee=tuple_annee+consommation(fichier)
```

```python
tuple_annee.sort()
x_dt=[]
y=[]
for element in tuple_annee:
        x_dt.append(element[0])
        y.append(element[1]/1000)
```

```python
init = x_dt[0]
t_init=time.mktime(init.timetuple())
```
